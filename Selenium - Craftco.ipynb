# Webscrapping 

## 2. Selenium - Craft Co

### 2.1. Extract Company (Name) Column For Further Webscraping 

```python
# !pip install selenium

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import TimeoutException
from time import sleep

import pandas as pd
import dbm
```

```python
#import file
cleaned_reddit_df = pd.read_csv('./beautifulsoup_reddit_data_cleaned.csv')
cleaned_reddit_df.head()
```

```python
company_df = cleaned_reddit_df["Company Name"]
company_list = company_df.values.tolist()
print(company_list)
```

### 2.2 Craft Co (Competitors)

```python
competitor_list = []

for company in company_list:
    driver = webdriver.Chrome('./chromedriver.exe') #need to download chromedriver into the same file 
    driver.get("https://craft.co/")
    elem = driver.find_element_by_name("q")
    elem.clear()
    sleep(3)
    elem.send_keys(company)
    sleep(3)
    
    elem.send_keys(Keys.RETURN)

    apply_button = WebDriverWait(driver, 100).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="app"]/div/div[2]/div/div[2]/ul/li[1]/a[2]')))
    apply_button.click() 
    sleep(3)

    competitor_button = WebDriverWait(driver, 100).until(EC.presence_of_element_located((By.LINK_TEXT, 'Competitors')))
    competitor_button.click()      
    sleep(3)

    competitors = WebDriverWait(driver,100).until(EC.presence_of_element_located((By.CLASS_NAME, 'compare-companies__chips')))
    competitors_name = competitors.text
    competitor_list.append(competitors_name)
    sleep(3)
    
    driver.quit()
    
print(competitor_list)
```

```python
#data clearning 
new_list = []

for competitor in competitor_list:
    cleaned_competitor = competitor.replace('\n',',')
    new_list.append(cleaned_competitor)

print(new_list)
```

```python
#store list of competitors into dataframe
df_craftco_competitors = pd.DataFrame(new_list)

#show dataframe
df_craftco_competitors.head()
```

```python
#split companies and their competitors into different columns
df_craftco_competitors_split = df_craftco_competitors.join(df_craftco_competitors[0].str.split(',', expand=True).add_prefix('competitor'))
df_craftco_competitors_split.head()
```

```python
#rename competitor0 as company
df_craftco_competitors_split.rename(columns={ df_craftco_competitors_split.columns[1]: "company" }, inplace = True)

#drop the redundant column
df_craftco_competitors_split = df_craftco_competitors_split.iloc[: , 1:]

#show dataframe
df_craftco_competitors_split.head()
```

```python
#save dataframe into csv file
df_craftco_competitors_split.to_csv('./selenium_competitors_data_cleaned.csv', sep=',', index = False) # Use Tab to separate
```
