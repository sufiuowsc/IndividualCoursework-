```python
#extract data from twitter API
empty_dict = {}
for index,query in enumerate(queries):
    empty_dict[query] = []
    cursor_test = tweepy.Cursor(api.search, q=query, tweet_mode="extended", lang ='en').items(100)
    stream_test = tweepy.Stream(auth, tweepy.StreamListener)

    for i in cursor_test:
        empty_dict[query].append(i.full_text)

#show dictionary with contains extracted tweets
print(empty_dict)
```

```python
#transform dictionary into dataframe
df_twitter = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in empty_dict.items()]))
df_twitter.head()
```
